####Conceptos


Minería de Datos Es el proceso técnico y automatizado de explorar grandes volúmenes de datos históricos para descubrir patrones repetitivos, correlaciones ocultas o anomalías estadísticas que permiten describir y estructurar la información para entender qué sucedió en el pasado.

Ciencia de Datos Es el campo multidisciplinario que combina matemáticas, estadística y programación para gestionar todo el ciclo de vida de la información (recolección, limpieza y análisis), con el objetivo de extraer conocimiento profundo y construir modelos predictivos que permitan anticipar el futuro y fundamentar la toma de decisiones.
